Численные методы решения алгебраических уравнений (2.3) 

упражнение 1

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

def f(x): 
    # Функция для метода простой итерации: φ(x) = 0.5*cos(x) + 1
    return 0.5 * np.cos(x) + 1


fig, ax = plt.subplots()

# Строим график функции и прямую y=x
x = np.linspace(0, 2, 100)
ax.plot(x, f(x), 'g-', label='φ(x)')
ax.plot(x, x, 'r--', label='y=x')
ax.grid(True, alpha=0.3)

# Метод простой итерации 
x0 = 0.5  # Начальное приближение
x_points = [x0]

for i in range(6):
    x_next = f(x_points[-1])  # Вычисляем следующее приближение
    
    # вертикаль к φ(x), затем горизонталь к y=x
    ax.plot([x_points[i], x_points[i]], [x_points[i], x_next], 'g-')  # вертикальна линия
    ax.plot([x_points[i], x_next], [x_next, x_next], 'g-')  # горизонтальная линия
    ax.plot(x_points[i], f(x_points[i]), 'ro', markersize=6)  # отмечаем точку итерации
    
    x_points.append(x_next)


ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()
ax.set_xlim(0, 2)
ax.set_ylim(0, 2)
ax.set_aspect('equal')


sns.set_style("whitegrid")
sns.set_context("notebook")

plt.show()


print("Итерации:", [f"{i}: {val:.6f}" for i, val in enumerate(x_points)])
```

```python
import numpy as np
from tabulate import tabulate

print("Решение уравнения x·sin(x) - 1 = 0")
print("Метод простой итерации, точность 10^-5")


# Итерационная функция: уравнение приведено к виду x = 1/sin(x)
def phi(x):
    return 1 / np.sin(x)

x = 1  # Начальное приближение
results = []  

# Итерационный процесс 
for n in range(1, 20):
    x_new = phi(x)  #  следующее приближение
    results.append([n, x_new, abs(x_new - x)])  # сохраняем результат
    
    # Проверка достижения заданной точности
    if abs(x_new - x) < 0.00001:
        
        headers = ["Итерация", "x", "|x_new - x|"]
        print(tabulate(results, headers=headers, floatfmt=".8f", tablefmt="grid"))
        print(f"Приближённое значение корня: {x_new:.8f}")
        print(f"Проверка: x * sin(x) - 1 = {x_new * np.sin(x_new) - 1:.8f}")
        break
    x = x_new  # Обновляем значение для следующей итерации
```
