задание 1

```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(-5, 5, 100)


y1 = 1.2*x**4 + 2*x**3 - 24.1
y2 = 13*x**2 + 14.2*x


plt.figure(figsize=(8, 6))
plt.plot(x, y1, 'b-', label='1.2x⁴ + 2x³ - 24.1')
plt.plot(x, y2, 'r-', label='13x² + 14.2x')
plt.grid(True)
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)
plt.xlabel('ось X')
plt.ylabel('ось Y')
plt.title('Графики функций')
plt.legend()


plt.show()



# Смотрим значения в разных точках
for i in range(len(x)-1):
    if (y1[i] - y2[i]) * (y1[i+1] - y2[i+1]) < 0:
        print(f"Пересечение между x = {x[i]:.2f} и x = {x[i+1]:.2f}")

```

задание 2

```python

def f(x):
    return 1.2*x**4 + 2*x**3 - 13*x**2 - 14.2*x - 24.1


a = 3.25
b = 3.5


epsilon = 0.001

# Счетчик итераций
iter_count = 0


print("Итерация |     a     |     b     |     x     |    f(x)   |  Длина интервала")
print("-" * 70)


while (b - a) > epsilon:
    iter_count += 1
    x = (a + b) / 2
    fx = f(x)
    
    
    print(f"{iter_count:^8} | {a:^9.6f} | {b:^9.6f} | {x:^9.6f} | {fx:^9.6f} | {b-a:^16.6f}")
    
    if fx == 0:
        break
    elif f(a) * fx < 0:
        b = x
    else:
        a = x


x_root = (a + b) / 2
print("-" * 70)
print(f"Корень уравнения: {x_root:.6f}")
print(f"Корень с точностью {epsilon}: {x_root:.3f}")
print(f"Количество итераций: {iter_count}")
```

```python

def g(x):

    return (( -2*x**3 + 13*x**2 + 14.2*x + 24.1) / 1.2) ** 0.25

# Начальное приближение
x = 3.2


epsilon = 0.000001


iter_count = 0

print("Итерация |     x     |    g(x)    |  |x - g(x)|")
print("-" * 45)


while True:
    iter_count = iter_count + 1
    x_new = g(x)
    diff = x_new - x
    if diff < 0:
        diff = -diff
    
    print(f"{iter_count:^8} | {x:^9.6f} | {x_new:^9.6f} | {diff:^11.6f}")
    
    if diff < epsilon:
        x = x_new
        break
    
    x = x_new

print("-" * 45)
print(f"Корень уравнения: {x:.6f}")
print(f"Корень с точностью 0.000001: {x:.6f}")
print(f"Количество итераций: {iter_count}")
```
