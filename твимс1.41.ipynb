задача 1

```python
import math
n = 100000
k = 5800
p = 0.06
q = 1 - p

# Вместо того чтобы считать n!, мы считаем ln(n!)
# math.lgamma(x) возвращает ln(Г(х))
# гамма-ф-ция связана с факториалом так: Г(n + 1) = n!
# следовательно ln(n!) = ln(Γ(n + 1)) = math.lgamma(n + 1)

# ln( n!/(k!(n-k)!) ) = ln(n!) - ln(k!) - ln((n-k)!)

log_comb = math.lgamma(n + 1) - math.lgamma(k + 1) - math.lgamma(n - k + 1)

# Вычисление логарифма вероятности 
# ln(p^k * q^(n-k)) = ln(p^k) + ln(q^(n-k)) = k * ln(p) + (n-k) * ln(q)

log_prob = k * math.log(p) + (n - k) * math.log(q)
log_result = log_comb + log_prob

# если мы сложили логарифмы, то это равносильно умножению чисел внутри логарифма - а это искомая вероятность P

prob = math.exp(log_result)
print(f"Вероятность = {prob:.10f}")
```

```python
import math

# Данные задачи
n = 1200
p = 0.002
q = 1 - p

 # Вычисляет P_n(k) через логарифмы
def bernoulli_log(n, k, p):
    q = 1 - p
# натуральный логарифм от числа сочетаний
    log_comb = math.lgamma(n + 1) - math.lgamma(k + 1) - math.lgamma(n - k + 1)
    log_prob = k * math.log(p) + (n - k) * math.log(q)
# Возвращает результат работы функции.
    return math.exp(log_comb + log_prob)


# Считаем вероятности для k = 0, 1, 2, 3
p0 = bernoulli_log(n, 0, p)
p1 = bernoulli_log(n, 1, p)
p2 = bernoulli_log(n, 2, p)
p3 = bernoulli_log(n, 3, p)


# Сумма вероятностей для k <= 3
summa = p0 + p1 + p2 + p3

# Искомая вероятность
ver = 1 - summa
print(f"Вероятность P(X > 3) = {ver:.6f}")
print(f"Вероятность P(X > 3) = {ver:.4%} (в процентах)")
```

```python
import math

# Данные задачи
n = 1000
p = 0.8
q = 1 - p

# Вычисляет логарифм P_n(k) через логарифмы
def bernoulli_log_ln(n, k, p):
    q = 1 - p
    log_comb = math.lgamma(n + 1) - math.lgamma(k + 1) - math.lgamma(n - k + 1)
    log_prob = k * math.log(p) + (n - k) * math.log(q)
    return log_comb + log_prob  # возвращаем логарифм, а не экспоненту



# Найдем вероятности в логарифмической форме для k = 0..14
log_probs = []
for k in range(15):
    log_p = bernoulli_log_ln(n, k, p) # получаем логарифм вероятности для данного k
    log_probs.append(log_p) # Добавляем это значение в список
    print(f"ln(P({k})) = {log_p:.2f}") 


max_log = max(log_probs) # максимальное значение среди всех логарифмов
sum_exp = 0
for log_p in log_probs:
    sum_exp += math.exp(log_p - max_log)

log_sum = max_log + math.log(sum_exp)
print(f"\nЛогарифм суммы вероятностей P(X < 15) = {log_sum:.2f}")
print(f"P(X < 15) ≈ e^{log_sum:.2f} = 10^{log_sum/math.log(10):.2f}")
```

Задача 7.
Вероятность того, что студент сдаст экзамен по высшей математике на "отлично", равна 0,25. Какова вероятность того, что из 40 студентов, сдающих экзамен, получат оценку "отлично" менее 8 студентов?

```python
import math

# Данные задачи
n = 40
p = 0.25  # вероятность сдать на "отлично"
q = 1 - p

# Вычисляет P_n(k) через логарифмы
def bernoulli_log(n, k, p):
    q = 1 - p
# натуральный логарифм от числа сочетаний
    log_comb = math.lgamma(n + 1) - math.lgamma(k + 1) - math.lgamma(n - k + 1)
    log_prob = k * math.log(p) + (n - k) * math.log(q)
# Возвращает результат работы функции.
    return math.exp(log_comb + log_prob)


# Считаем вероятности для k = 0, 1, 2, ..., 7 (менее 8, то есть от 0 до 7)
probabilities = []
for k in range(8):  # k = 0, 1, 2, ..., 7
    prob = bernoulli_log(n, k, p)
    probabilities.append(prob)
    print(f"P({k}) = {prob:.6f} ({prob:.2e})")


# Сумма вероятностей для k <= 7 (менее 8)
summa = sum(probabilities)

# Искомая вероятность (менее 8 отличников)
ver = summa
print(f"\nВероятность P(X < 8) = {ver:.6f}")
print(f"Вероятность P(X < 8) = {ver:.4%} (в процентах)")
```
