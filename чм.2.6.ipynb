### Упражнения 2.6 Решение алгебраических и трансцендентных уравнений с помощью инструментальных средств

упр 1

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.log10(x) - np.cos(x)

a, b = 0.1, 3
step = 0.01

interval = None
x = a

while x < b:
    if f(x) * f(x + step) < 0:
        interval = (x, x + step)
        break
    x += step

# Метод половинного деления (бисекции) 
eps = 1e-4
left, right = interval[0], interval[1]

while (right - left) / 2 > eps:
    c = (left + right) / 2
    if f(left) * f(c) < 0:
        right = c
    else:
        left = c

root = (left + right) / 2

x_vals = np.linspace(0.1, 3, 1000)
y_vals = np.log10(x_vals) - np.cos(x_vals)

plt.figure(figsize=(8,5))
plt.plot(x_vals, y_vals, label='lg(x) - cos(x)')
plt.axhline(0)

plt.scatter(root, 0, color='red', s=80)
plt.text(root, 0, f'{root:.4f}', color='red')

plt.grid()
plt.legend()
plt.title("Локализация корня")
plt.show()

print(f"Корень: x ≈ {root:.4f}")
```

упр 2

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2 - 3.2*x - 1

x = np.linspace(-2, 5, 1000)
y = f(x)

# Метод Ньютона (касательных)
x0 = 3
eps = 1e-4

while True:
    df = 2*x0 - 3.2  # производная функции f'(x) = 2x - 3.2
    x1 = x0 - f(x0)/df
    if abs(x1 - x0) < eps:
        root = x1
        break
    x0 = x1

plt.figure(figsize=(8,5))
plt.plot(x, y, label='x^2 - 3.2x - 1')
plt.axhline(0)

plt.scatter(root, 0, color='red', s=80)
plt.text(root, 0, f'{root:.4f}', color='red')

plt.grid()
plt.legend()
plt.title("Задание 2")
plt.show()

print(f"Корень: x ≈ {root:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Функции
f = lambda x: 4*x - np.log(x) - 4 if x > 0 else np.nan
phi = lambda x: (4 + np.log(x)) / 4

# Метод итераций
x = 2.0
while True:
    x_new = phi(x)
    if abs(x_new - x) < 10 ** -4:
        break
    x = x_new

# Визуализация
x_vals = np.linspace(0.1, 3, 300)
plt.plot(x_vals, [f(v) for v in x_vals], 'b-')
plt.axhline(0, c='r', ls='--')
plt.plot(x_new, 0, 'ro', label=f'x = {x_new:.4f}')
plt.grid(alpha=0.3)
plt.title('4x - ln x = 4')
plt.legend()
plt.show()

print(f'Корень: {x_new:.4f}')
```
