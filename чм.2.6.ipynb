упр 2

```python
import numpy as np
import matplotlib.pyplot as plt

# Исходная функция
def func(x):
    return x**2 - 3.2*x - 1

# Приводим к виду x = φ(x) для метода итераций
def phi(x):
    return (x**2 - 1) / 3.2

# Графическое отделение корней
x = np.linspace(-2, 5, 300)
plt.plot(x, func(x), 'b-', label='f(x)')
plt.plot(x, x*0, 'r--', label='y=0')
plt.grid(alpha=0.3)
plt.title('Локализация корней')
plt.legend()
plt.show()

# Метод простой итерации
x0 = 3.3  # начальное приближение
eps = 1e-4
diff = 1
iter_count = 0

while diff > eps:
    x1 = phi(x0)
    diff = abs(x1 - x0)
    x0 = x1
    iter_count += 1

print(f'Корень: {x1:.4f}')
print(f'Итераций: {iter_count}')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Функции
f = lambda x: 4*x - np.log(x) - 4 if x > 0 else np.nan
phi = lambda x: (4 + np.log(x)) / 4

# Метод итераций
x = 2.0
while True:
    x_new = phi(x)
    if abs(x_new - x) < 1e-4:
        break
    x = x_new

# Визуализация
x_vals = np.linspace(0.1, 3, 300)
plt.plot(x_vals, [f(v) for v in x_vals], 'b-')
plt.axhline(0, c='r', ls='--')
plt.plot(x_new, 0, 'ro', label=f'x = {x_new:.4f}')
plt.grid(alpha=0.3)
plt.title('4x - ln x = 4')
plt.legend()
plt.show()

print(f'Корень: {x_new:.4f}')
```
